```python
cmu_plot = pd.read_table('plot_summaries.txt', header = None)
cmu_plot.columns = ['ID', 'plot']
cmu_plot.head()
```

```python
cmu = pd.read_table('movie.metadata.tsv', header=None)
cmu2 = cmu[[2,3]]
cmu2[2] = cmu2[2].apply(lambda x: x.lower())
cmu2[3] = cmu2[3].fillna('0')
cmu2[4] = cmu2[3].apply(lambda x: len(x))
cmu2.loc[cmu2[4]>4, 3] = cmu2.loc[cmu2[4]>4, 3].apply(lambda x: x[:4])
cmu2 = cmu2[[2,3]]
cmu2.columns = ['Title', 'Release Year']
cmu2['Release Year'] = cmu2['Release Year'].astype(int)
```
